`Video Game Sales Dataset`
1. **Rank**: This column likely represents the ranking or position of the game in the dataset. It indicates how the game is ranked in terms of sales or popularity, with lower numbers indicating higher ranks.

2. **Name**: This column contains the name of the video game. It's the title by which the game is commonly known.

3. **Platform**: This column specifies the gaming platform or console on which the game was released. Examples of platforms include PlayStation, Xbox, Nintendo Switch, PC, etc.

4. **Year**: This column shows the year in which the game was released. It's the year when the game became available to the public.

5. **Genre**: This column indicates the genre or category to which the game belongs. Examples of genres include action, adventure, sports, role-playing, strategy, etc.

6. **Publisher**: This column lists the name of the company or entity that published or released the game. The publisher is responsible for marketing, distribution, and sometimes development of the game.

7. **NA_Sales**: This column represents the sales of the game in North America (NA), often measured in millions of copies or units sold.

8. **EU_Sales**: This column represents the sales of the game in Europe (EU), again usually measured in millions of copies or units sold.

9. **JP_Sales**: This column represents the sales of the game in Japan (JP), measured in millions of copies or units sold. Japanese sales figures are often important, as the gaming market in Japan has unique trends.

10. **Other_Sales**: This column includes sales figures from regions other than North America, Europe, and Japan. These regions could include countries in other parts of the world, such as Asia, Australia, and South America.

11. **Global_Sales**: This column represents the total worldwide sales of the game, which is the sum of sales in North America, Europe, Japan, and other regions.

There are `271` Missing values in **Year** Column and `58` in **Publisher** Column.

- **`Handling the null the values of Year Column`**

```python
# Fill all the null values by its mean
df['Year'].fillna(df['Year'].mean(),inplace = True)
```

```python
df['Year'] = np.ceil(df['Year'])
df['Year'] = df['Year'].astype(int)
```

# Conclusion

Based on the analysis of the video game sales dataset, several key insights can be drawn:

1. **Platform Trends**: The platform with the most game releases is PS2, followed closely by DS. These platforms have been prolific in terms of hosting a wide range of games.

2. **Release Years**: The years 2007, 2008, and 2009 saw the highest number of game releases. However, the number of releases significantly declined after 2009, indicating a potential shift in the industry.

3. **Genre Popularity**: The action genre has the highest number of game releases, followed by sports. Action games seem to dominate the market in terms of sheer numbers.

4. **Publisher Impact**: Electronic Arts emerges as the leading publisher with the most games released. This suggests its strong presence and influence in the gaming industry.

5. **Global Sales Distribution**: Distribution plots show that global sales tend to be concentrated in the lower range for all regions (NA, EU, JP, Other, and Global). However, a few games have exceptionally high global sales.

6. **Platform Performance**: Certain platforms like the PS2 and DS have seen impressive global sales, indicating their popularity among gamers.

7. **Genre and Global Sales**: The action genre consistently boasts the highest global sales, followed by sports and shooter genres. These genres seem to resonate well with players on a global scale.

8. **Yearly Global Sales**: Over the years, global game sales have experienced a peak between 2006 and 2009, with a subsequent decline in sales. This suggests a potential shift in gaming trends.

In conclusion, the analysis provides valuable insights into the dynamics of the video game industry. The data highlights the dominance of certain genres, platforms, and publishers, while also showcasing trends in global sales. Understanding these patterns can aid in strategic decision-making and offer insights into the preferences of gamers and the evolving landscape of the gaming market.
